```python
import pandas as pd
import os
```

# Merging the 12 months of sales data into a single csv file

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")

files = [file for file in os.listdir('./Sales_data')]

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data,df])

all_months_data.to_csv('Merge_data.csv', index = False)
```

# Read data in updated dataframe

```python
all_data = pd.read_csv('Merge_data.csv')
all_data.head()
```

### Clean up the data! basically removing the nan values

### Find all nan values and remove them

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

```python
# to find and resolve the 'Or' issue we face while converting month column to int: invalid literal for int() with base 10: 'Or'
#temp_data = all_data[all_data['Order Date'].str[0:2] == 'Or']
#temp_data.head()

#resolving the issue
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

### convert columns to corect format to solve this issue: can't multiply sequence by non-int of type 'str'

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
all_data.head()
```

### Augment data with additional columns

### Task 2: Add month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Question 1: What was the best month for sales? how muh was earned that month?

```python
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales')
plt.xlabel('month number')
plt.show()
```

### Question 2: which city had the highest number of sales?

### Add a city column

```python
# using .apply() of pandas. .apply() method helps to run a random function in pandas

def get_state(address):
    return address.split(',')[2].split()[0]

def get_city(address):
    return address.split(',')[1]

# all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")

all_data.head()
```

```python
results_city = all_data.groupby('City').sum()
results_city
```

```python
import matplotlib.pyplot as plt

#cities = all_data['City'].unique() ## createsproblem with the order
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results_city['Sales'])
plt.xticks(cities, rotation = 'vertical', size = 8)
plt.ylabel('Sales')
plt.xlabel('city')
plt.show()
```

### At what time should we advertise to maximize the likelihood of customer's buying products?

## Augmenting data into date and min column

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]
hours, all_data.groupby('Hour').count()
plt.plot(hours, all_data.groupby('Hour').count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Number of orders')
plt.grid()
plt.show()
```

### What products are most often sold together?

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

#print(count)

for key, value in count.most_common(10):
    print(f"{key} : {value}")
```
